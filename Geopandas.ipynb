# Ubicación geográfica de los eventos

Geopandas permite leer archivos shapefile (.shp), que contiene informacion geografica.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
```

## Ubicacion de los archivos necesarios

```python
folderInputPath = '../shapefileInput/'
inputListPath = [folderInputPath + path + '/' for path in os.listdir(folderInputPath)]
inputListPath
```

```python
inputPath_zca = inputListPath[1]
inputPath_pa  = inputListPath[2]
inputPath_dep = inputListPath[5]

fileListPath_zca = [file for file in os.listdir(inputPath_zca)]
fileListPath_pa  = [file for file in os.listdir(inputPath_pa)]
fileListPath_dep = [file for file in os.listdir(inputPath_dep)]
```

```python
fileListPath_zca
```

```python
fileListPath_pa
```

```python
fileListPath_dep
```

## Lectura y observacion de los archivos

```python
zonasX_zca = gpd.read_file(inputPath_zca + fileListPath_zca[7])
zonasX_zca.head()
```

```python
zonasX_pa  = gpd.read_file(inputPath_pa + fileListPath_pa[8])
zonasX_pa.iloc[:, [2, 3, 16, 17, 28]]
```

```python
zonasX_dep = gpd.read_file(inputPath_dep + fileListPath_dep[2])
zonasX_dep.head()
```

## Verificamos que los CRS esten en epsg:32618

```python
zonasX_pa.crs
```

```python
zonasX_dep.crs
```

```python
zona2_pa = zonasX_pa.to_crs({'init': 'epsg:32618'})
```

```python
zona2_dep = zonasX_dep.to_crs({'init': 'epsg:32618'})
```

## Seleccionamos la region Ayacucho

```python
ayacucho = gpd.GeoDataFrame(zona2_dep.iloc[4:5])
```

```python
zona2_dep.plot(cmap = 'brg', column='IDDPTO', figsize=(12, 12))
```

## Graficamos la region junto con los puntos donde se encuentran los eventos

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
ayacucho.plot(ax=ax, cmap = 'Set3')
zona2_pa.plot(ax=ax, cmap = 'jet')
```
